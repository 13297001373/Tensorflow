# MNIST数据集分类（手写体识别）

```python
import warnings

warnings.filterwarnings('ignore')
```

```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
```

```python
##加载数据
mnist = input_data.read_data_sets('MNIST_data',one_hot=True)

##设置每一批次的数据个数
batch_size = 100

##确定批次的个数
n_batch = mnist.train.num_examples // batch_size

##定义x,y的placeholder
x = tf.placeholder(tf.float32,[None,28*28])
y = tf.placeholder(tf.float32,[None,10])


##创建简单神经网络
W = tf.Variable(tf.random_normal([28*28,10]))
b = tf.Variable(tf.random_normal([10]))
result = tf.nn.softmax(tf.matmul(x,W) +b)


##1.定义损失函数
# loss = tf.reduce_mean(tf.square(result-y))  ##二次代价损失函数
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y,logits=result))  ##采用交叉熵使用

##采用dropout来防止过拟合1

##2.设置优化算法
optimizer = tf.train.GradientDescentOptimizer(0.1)

##3.使用优化算法减小loss
train = optimizer.minimize(loss)

##初始化变量
init = tf.global_variables_initializer()


##定义评判方式(bool型的列表中)
correct_prediction = tf.equal(tf.argmax(y,1),tf.argmax(result,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction,tf.float32))


##启动会话
with tf.Session() as sess:
    sess.run(init)
    
    for epoch in range(201):  ##定义迭代次数
        for batch in range(n_batch):
            batch_xs,batch_ys = mnist.train.next_batch(batch_size)
            sess.run(train,feed_dict={x:batch_xs,y:batch_ys})
        print('第',epoch+1,'轮，训练集准确率为：',sess.run(accuracy,feed_dict={x:batch_xs,y:batch_ys}))
    print('\n测试准确率为：',sess.run(accuracy,feed_dict={x:mnist.test.images,y:mnist.test.labels})) 
```
