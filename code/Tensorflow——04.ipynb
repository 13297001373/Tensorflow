# 4-1 dropout的使用

```python
import warnings

warnings.filterwarnings('ignore')
```

```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
```

```python
##加载数据
mnist = input_data.read_data_sets('MNIST_data',one_hot=True)

##设置每一批次的数据个数
batch_size = 100

##确定批次的个数
n_batch = mnist.train.num_examples // batch_size

##定义x,y的placeholder
x = tf.placeholder(tf.float32,[None,28*28])
y = tf.placeholder(tf.float32,[None,10])

##此时加入dropout的比例
keep_prob = tf.placeholder(tf.float32)


##创建第一层神经网络
W1 = tf.Variable(tf.truncated_normal([28*28,2000],stddev=0.1))  #没有选择0初始化
b1 = tf.Variable(tf.zeros([2000])+0.1)
Z1 = tf.matmul(x,W1)+b1
A1 = tf.nn.tanh(Z1)  ##激活函数选用tanh
A1_drop = tf.nn.dropout(A1,keep_prob=keep_prob)  #dropout


##创建输出层神经网络
W2 = tf.Variable(tf.truncated_normal([2000,10],stddev=0.1))
b2 = tf.Variable(tf.zeros([10])+0.1)
Z2 = tf.matmul(A1_drop,W2) + b2
result = tf.nn.softmax(Z2)


##1.定义损失函数
# loss = tf.reduce_mean(tf.square(result-y))  ##二次代价损失函数
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y,logits=result))  ##采用交叉熵使用


##2.设置优化算法
optimizer = tf.train.GradientDescentOptimizer(0.1)

##3.使用优化算法减小loss
train = optimizer.minimize(loss)

##初始化变量
init = tf.global_variables_initializer()


##定义评判方式(bool型的列表中)
correct_prediction = tf.equal(tf.argmax(y,1),tf.argmax(result,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction,tf.float32))


##启动会话
with tf.Session() as sess:
    sess.run(init)
    
    for epoch in range(201):  ##定义迭代次数
        for batch in range(n_batch):
            batch_xs,batch_ys = mnist.train.next_batch(batch_size)
            sess.run(train,feed_dict={x:batch_xs,y:batch_ys,keep_prob:1.0})
        print('第',epoch+1,'轮，训练集准确率为：',sess.run(accuracy,feed_dict={x:batch_xs,y:batch_ys,keep_prob:1.0}))
    print('\n测试准确率为：',sess.run(accuracy,feed_dict={x:mnist.test.images,y:mnist.test.labels,keep_prob:1.0})) 
```
